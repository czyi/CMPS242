```python
from PIL import Image
import matplotlib.pyplot as plt
img=Image.open('pixiv/part_01_ori/2007_10_29_21_00_09_126025_p0.jpg')  #打开图像

scale = 256
halfs = 128

print(img.size)

x = img.size[0]//scale
print(x)
y = img.size[1]//scale
print(y)

plt.figure("beauty")
plt.subplot(1,2,1), plt.title('origin')
plt.imshow(img),plt.axis('off')

xx = img.size[0]//2-halfs*x
yy = img.size[1]//2-halfs*y
img_num = 0
for i in range(x):
    for j in range(y):
#         box=(img.size[0]//2-halfs*x,img.size[1]//2-halfs*y,img.size[0]//2+halfs*x,img.size[1]//2+halfs*y)
        box = (xx+i*scale, yy+j*scale, xx+(i+1)*scale, yy+(j+1)*scale)
        print(box)
        roi=img.crop(box)
        roi.save('pixiv/part_01/'+str(img_num)+'.jpg')
        img_num += 1


plt.subplot(1,2,2), plt.title('roi')
plt.imshow(roi),plt.axis('off')
plt.show()
```

```python
from PIL import Image
import os
import numpy as np
```

```python
scale = 400
halfs = 200
img_num = 589

folder = 'pixiv/part_02_sel'
for img_file in os.listdir(folder):
    img = Image.open(folder+'/'+img_file)
    
    print(img.size)
#     print(img.mode)
    if(img.mode!='RGB'):
        continue
    x = img.size[0]//scale
    y = img.size[1]//scale

    xx = img.size[0]//2-halfs*x
    yy = img.size[1]//2-halfs*y
    for i in range(x):
        for j in range(y):
    #         box=(img.size[0]//2-halfs*x,img.size[1]//2-halfs*y,img.size[0]//2+halfs*x,img.size[1]//2+halfs*y)
            box = (xx+i*scale, yy+j*scale, xx+(i+1)*scale, yy+(j+1)*scale)
#             print(box)
            roi=img.crop(box)
            roi_arr = np.reshape(np.array(roi), scale*scale*3)
            if((np.sum(roi_arr>=250)/np.sum(roi_arr>=0)) > 0.5):
                print(np.sum(roi_arr==255)/np.sum(roi_arr>=0))
                continue
        
            roi.save('pixiv/part_02/'+str(img_num)+'.jpg')
            img_num += 1
            
```
